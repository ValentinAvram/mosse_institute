```python
import pandas as pd

def group_columns(df, col1, col2, new_col_name):
    df[new_col_name] = df[col1].astype(str) + ' -> ' + df[col2].astype(str)
    return df


new_df = group_columns(df, 'name', 'pid', 'PID-NAME')

display(new_df)
```
